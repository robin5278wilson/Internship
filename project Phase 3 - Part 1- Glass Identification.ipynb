##### Here the boxplot for BA:BARIUM column seems to contribute nothing to the target column as its a flat box plot meaning all its value in the quartiles are the same or very close to each other. The same is also confirmed using the describe() method. This suggests that the column must be dropped for good. But before doing so i want to check the outliers in the column

```python
# outliers in barium column
q1=dfg['Ba: Barium'].quantile(0.25)
q3=dfg['Ba: Barium'].quantile(0.75)
iqr=q3-q1
lower=q1-1.5*iqr
upper=q3+1.5*iqr
outliers=dfg[(dfg['Ba: Barium']<lower)|(dfg['Ba: Barium']>upper)]
outliers
```

```python
#Checking the count of each category in target column
sns.countplot(x=dfg['Type of glass'],data=dfg)
print(dfg['Type of glass'].value_counts())
```

```python
#creating new target column with only 2 classes. Glass used for window is TRUE and rest is false
dfg['New Target']=dfg['Type of glass']<=4
```

```python
# Converting new target into 0 and 1
dfg['New Target']=dfg['New Target'].astype(int)
```

```python
df1=dfg.drop('Type of glass', axis=1,)
```

```python
# Dropping the ID and Ba: Barium columns as they have no contribution to the target
df1.drop(['Id number','Ba: Barium'],axis=1,inplace=True)
```

```python
#Checking count of new target column
print(df1['New Target'].value_counts())
sns.countplot(x=df1['New Target'],data=df1)
```

```python
# removing outliers using z-score
from scipy import stats

outlier_col = ['RI: refractive index', 'Na: Sodiumin', 'Mg: Magnesium', 'Al: Aluminum', 'Si: Silicon', 'K: Potassium', 'Ca: Calcium', 'Fe: Iron']

z_scores = stats.zscore(df1[outlier_col])
z_scores_df = pd.DataFrame(z_scores, columns=outlier_col)

threshold = 3

df_cleaned = df1[(z_scores_df < threshold).all(axis=1)]

# Check the new class distribution for df_cleaned
print(df_cleaned['New Target'].value_counts())
```

```python
plt.figure(figsize=(15, 10))

# Generate a box plot for each column
for i, column in enumerate(df_cleaned.columns, 1):
    plt.subplot(3, 4, i)
    sns.boxplot(y=df_cleaned[column])
    plt.title(column)

# Adjust the layout
plt.tight_layout()
plt.show()
```

```python
# Creating x anf y
x = df_cleaned.drop('New Target', axis=1)
y = df_cleaned['New Target']
```

```python
# performing SMOTE to get balanced dataset with equally distributed target categories
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
x_smote, y_smote = smote.fit_resample(x, y)

# Check the new class distribution
print(pd.Series(y_smote).value_counts())
```

```python
x_smote
```

```python
#Scaling the data using standard scaler
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
x_scaled=scale.fit_transform(x_smote)
```

```python
xscaled=pd.DataFrame(x_scaled)
```

```python
xscaled
```

```python
# Splitting data into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(xscaled,y_smote,test_size=0.20,random_state=44)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(max_depth= None, min_samples_split= 2, n_estimators= 50)
rf.fit(x_train,y_train)
```

```python
rf_pred=rf.predict(x_test)
```

```python
from sklearn.metrics import classification_report, accuracy_score
print('Accuracy score:',accuracy_score(y_test,rf_pred))
```

```python
print('Confusion Matrix :')
print(classification_report(y_test,rf_pred))
```
