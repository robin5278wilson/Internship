```python
# creating empty list to store company names
com=[]
```

```python
# Extracting first 10 company names
company=driver.find_elements(By.XPATH,'//div[@class=" row2"]/span/a[1]')
for i in company[0:10]:
    com.append(i.text)
com
```

```python
# creating empty list to store job location
location=[]
```

```python
# creating empty list to store job experience
exp=[]
```

```python
# Extracting Job Experience
experience=driver.find_elements(By.XPATH,'//span[@class="exp-wrap"]/span/span')
for i in experience[0:10]:
    exp.append(i.text)
exp
```

```python
# Finding lengths before creating dataframe
print(len(jbt),len(com),len(location),len(exp))
```

```python
# Creating dataframe using the extracted values
naukri_jobs=pd.DataFrame({'Job Title':jbt,'Company Name':com,'Experience required':exp,'Job Location':location})
naukri_jobs
```

# Solving Question 7

```python
#Automating chrome
drv=webdriver.Chrome()
drv.get('https://www.jagranjosh.com/general-knowledge/list-of-all-prime-ministers-of-india-1473165149-1')
```

```python
#Extracting Headers from table
columns=[]
tab_cols=drv.find_elements(By.XPATH,'//div[@class="TableData"]/table/tbody/tr/th')
for i in tab_cols:
    columns.append(i.text)
columns
```

```python
len(columns)
```

```python
# Extracting entire table content
rows=[]

for i in range(2,21):
    xpath='//div[@class="TableData"]/table/tbody/tr['+str(i)+']/td'
    tab_rows=drv.find_elements(By.XPATH,xpath)
    for i in tab_rows:
        rows.append(i.text)
rows
```

```python
# Converting extracted table into array
import numpy as np
tabrow=np.array(rows)
```

```python
tabrow
```

```python
# Reshaping array into a 19row 5 column format
array=tabrow.reshape(19,5)
```

```python
array
```

```python
# Creating Dataframe
table_data=pd.DataFrame(data=array,columns=columns)
```

```python
table_data
```
