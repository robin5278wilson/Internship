### From this we can understand that there are only 13 rows (571-558) where a row has more that 1 nan value. Deleting such rows make more sense as this way we will not be deleting 154 rows as seen in line cell no 85.

```python
# deleting rows with more than 1 nan values
dfnew=df[df.isnull().sum(axis=1)<=1]
```

```python
dfnew.isnull().sum()
```

#### Now there are a total of 141 nan values in only 3 columns.  

```python
dfnew[dfnew.isnull().any(axis=1)]
```

```python
# creating list of columns to replace nan values with mode
nancols=['CS-406','CS-423','CS-412']
```

```python
# replacing nan values with mode from the respecting columns
for i in dfnew.columns:
    mode_value = dfnew[i].mode()[0]
    dfnew[i].fillna(mode_value, inplace=True)
```

```python
dfnew.isnull().sum()
```

### Now all nan values are removed and replaced with mode

```python
x.columns
```

```python
# Since the x has ordinal categorical data values hence we will use ordinal encoder to encode the same
from sklearn.preprocessing import OrdinalEncoder
grade_order = [['F', 'D', 'D+', 'C-', 'C', 'C+', 'B-', 'B', 'B+', 'A-', 'A', 'A+']]
encoder=OrdinalEncoder(categories=grade_order*len(x.columns))
x_encoded=encoder.fit_transform(x)
```

```python
xnew=pd.DataFrame(x_encoded)
```

```python
xnew
```

#### Now since the encoding is done, lets proceed to train_test_split and model building, fitting and predictions

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

x_train,x_test,y_train,y_test=train_test_split(xnew,y,test_size=0.2,random_state=39)
```

```python
print('len X-train:',len(x_train))
print('len Y-train:',len(y_train))
print('len X-test:',len(x_test))
print('len Y-test:',len(y_test))
      
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
lrpred=lr.predict(x_test)
```

```python
print(mean_squared_error(y_test,lrpred))
```

```python
print("R2 Score : ",r2_score(y_test,lrpred))
```
