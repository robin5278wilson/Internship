```python
# Checking for outliers using Box plot
for i,col in enumerate(var.columns):
    plt.figure(figsize=(10, 15))
    plt.subplot(len(var.columns),2,i+1)
    sns.boxplot(df[col])
    plt.title('Box Plot')
    plt.xlabel(col)
```

```python
#Based on the boxplot i have identified the below columns which needs to be treated for outliers
outlier_cols=['Family','Trust (Government Corruption)','Generosity','Dystopia Residual', 'Standard Error']
```

```python
df.shape
```

```python
# Treating outliers using the quantile/IQR method
q1=df[outlier_cols].quantile(0.25)
q3=df[outlier_cols].quantile(0.75)
iqr=q3-q1
lower=q1-1.5*iqr
upper=q3+1.5*iqr
outliers=var[((df[outlier_cols]<lower)|(df[outlier_cols]>upper)).any(axis=1)]
newvar=var[~((df[outlier_cols]<lower)|(df[outlier_cols]>upper)).any(axis=1)]
newvar
```

#### We can see that 32 rows with outliers were removed from dataset

```python
for i,col in enumerate(newvar.columns):
    plt.figure(figsize=(10, 25))
    plt.subplot(len(newvar.columns),1,i+1)
    sns.boxplot(y=newvar[col])
    plt.title('Box Plot')
    plt.xlabel(col)
```

### Outliers has been taken care off to an extent.

```python
# Seperating x data
x=newvar.drop(['Happiness Score'],axis=1)
```

```python
x
```

#### Also since standard error is highly negatively correlated to Happiness score, as per heatmap so we will drop it

```python
x=x.drop(['Standard Error'],axis=1)
```

```python
x
```

```python
# Determining target label
y=newvar['Happiness Score']
```

```python
y
```

```python
# Checking the dtype before proceeding
x.info()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
pred=lr.predict(x_test)
```

```python
# Checking performance of the model
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae = mean_absolute_error(y_test, pred)
mse = mean_squared_error(y_test, pred)
rmse = mean_squared_error(y_test, pred, squared=False)  # squared=False to get RMSE
r2 = r2_score(y_test, pred)

print(f'Mean Absolute Error (MAE): {mae}')
print(f'Mean Squared Error (MSE): {mse}')
print(f'Root Mean Squared Error (RMSE): {rmse}')
print(f'R-squared (R²) Score: {r2}')
```

## The r2 score is very high at 99.99%. Suggesting that the model is accurately predicting happiness scores

```python
# Using cross validation to recheck model performance
from sklearn.model_selection import cross_val_score
scores = cross_val_score(lr, x_train, y_train, cv=5, scoring='r2')

print(f'Cross-validated R² scores: {scores}')
print(f'Average R² score: {scores.mean()}')
```
