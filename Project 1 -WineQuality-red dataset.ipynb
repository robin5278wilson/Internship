```python
# Define a threshold for outlier detection using IQR method
threshold = 1.5

# Iterate over each column and apply outlier removal
for col in cols:
    Q1 = dwcopy[col].quantile(0.25)
    Q3 = dwcopy[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - threshold * IQR
    upper_bound = Q3 + threshold * IQR
    dwcopy = dwcopy[(dwcopy[col] >= lower_bound) & (dwcopy[col] <= upper_bound)]
```

```python
# STandardizing the input data
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
x_scaled=scale.fit_transform(dwcopy)
x_data=pd.DataFrame(data=x_scaled,columns=dwcopy.columns)
```

```python
x_data
```

```python
#Spliting data into train and test
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score

x_train,x_test,y_train,y_test=train_test_split(x_data,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
# Hyperparameter tuning with the free sulfur dioxide column
import warnings
warnings.filterwarnings('ignore')
models = {'RandomForest': RandomForestClassifier(),'GradientBoosting': GradientBoostingClassifier(),'SVM': SVC()}
params = {
    'RandomForest': {
        'n_estimators': [100, 200],
        'max_depth': [10, 20],
        'min_samples_split': [2, 5],
        'min_samples_leaf': [1, 2],
        'max_features': ['sqrt'],
        'bootstrap': [True],
        'criterion': ['gini']
    },
    'GradientBoosting': {
        'n_estimators': [100, 200],
        'learning_rate': [0.01, 0.1],
        'max_depth': [3, 5],
        'min_samples_split': [2, 5],
        'min_samples_leaf': [1, 2],
        'subsample': [0.7, 0.8],
        'max_features': ['sqrt']
    },
    'SVM': {
        'C': [0.1, 1, 10],
        'kernel': ['linear', 'rbf'],
        'gamma': ['scale']
    }
}
best_models = {}
for model in models:
    grid_search = GridSearchCV(models[model], params[model], cv=5, scoring='accuracy', n_jobs=-1)
    grid_search.fit(x_train, y_train)
    best_models[model] = grid_search.best_estimator_
    print(f"Best parameters for {model}: {grid_search.best_params_}")

for model_name in best_models:
    model = best_models[model_name]
    y_pred = model.predict(x_test)
    print(f"\n{model_name} Classification Report:")
    print(classification_report(y_test, y_pred))
    print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
```

```python
from sklearn.linear_model import LogisticRegression
lmodel=LogisticRegression()
```

```python
lparams = {
    'penalty': ['l1','l2','elasticnet'],
    'C': [0.01, 0.1, 1, 10, 100],
    'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],
    'max_iter': [100, 200, 300, 500],
    'class_weight': [None, 'balanced']
}

# Perform Grid Search
grid_search = GridSearchCV(estimator=lmodel, param_grid=lparams, cv=5, scoring='accuracy', n_jobs=-1, verbose=2)
grid_search.fit(x_train, y_train)

# Best model and parameters
best_model = grid_search.best_estimator_
print(f"Best parameters: {grid_search.best_params_}")

# Evaluate Best Model
y_pred = best_model.predict(x_test)
print("\nLogistic Regression Classification Report:")
print(classification_report(y_test, y_pred))
print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
```

NOTE:
Looking at the above 2 hyperparameter tuning results and looking at the accuracy score and classification report we can conclude that the GradientBoosting model was able to perform the best as it predicts the class 1 data more accurately than the other models. The data however is still imbalanced.

```python
# Will try using the smote technique on the dataset to see if it performs better
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
x_resampled, y_resampled = smote.fit_resample(x_data, y)
```

```python
# Splitting data into train and test
x_retrain,x_retest,y_retrain,y_retest=train_test_split(x_resampled,y_resampled,test_size=0.2,random_state=42)
```

```python
# Hyperparameter tuning on the SMOTE resampled data
import warnings
warnings.filterwarnings('ignore')
models = {'RandomForest': RandomForestClassifier(),'GradientBoosting': GradientBoostingClassifier(),'SVM': SVC()}
params = {
    'RandomForest': {
        'n_estimators': [100, 200],
        'max_depth': [10, 20],
        'min_samples_split': [2, 5],
        'min_samples_leaf': [1, 2],
        'max_features': ['sqrt'],
        'bootstrap': [True],
        'criterion': ['gini']
    },
    'GradientBoosting': {
        'n_estimators': [100, 200],
        'learning_rate': [0.01, 0.1],
        'max_depth': [3, 5],
        'min_samples_split': [2, 5],
        'min_samples_leaf': [1, 2],
        'subsample': [0.7, 0.8],
        'max_features': ['sqrt']
    },
    'SVM': {
        'C': [0.1, 1, 10],
        'kernel': ['linear', 'rbf'],
        'gamma': ['scale']
    }
}
best_models = {}
for model in models:
    grid_search = GridSearchCV(models[model], params[model], cv=5, scoring='accuracy', n_jobs=-1)
    grid_search.fit(x_retrain, y_retrain)
    best_models[model] = grid_search.best_estimator_
    print(f"Best parameters for {model}: {grid_search.best_params_}")

for model_name in best_models:
    model = best_models[model_name]
    y_repred = model.predict(x_retest)
    print(f"\n{model_name} Classification Report:")
    print(classification_report(y_retest, y_repred))
    print(f"Accuracy: {accuracy_score(y_retest, y_repred)}")
```

Hence we can see that the performance of the model has gone up by around 3% for randomforest and gradientboosting but has gone up considerably by 5.5% for SVM. However still the gradient boosting model performs the best.
