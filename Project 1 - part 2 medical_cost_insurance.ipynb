```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_scaled=scaler.fit_transform(med_x)
```

```python
x_scaled
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,med_y,test_size=0.20,random_state=42)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import GridSearchCV

gbr = GradientBoostingRegressor()

# hyperparameters tuning inputs
gbrparams = {
    'n_estimators': [50, 100, 250],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [4, 6, 8],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 5],
    'subsample': [0.5, 0.7, 0.8, 0.9],
    'max_features': ['sqrt', 'log2']
}

# Use GridSearchCV for hyperparameter tuning
grid_search = GridSearchCV(estimator=gbr, param_grid=gbrparams, cv=5, scoring='neg_mean_squared_error', n_jobs=-1)
grid_search.fit(x_train, y_train)

# Get the best model
best_gbr = grid_search.best_estimator_

# Print the best parameters
print("Best parameters found: ", grid_search.best_params_)
```

```python
boost=GradientBoostingRegressor(learning_rate=0.1, max_depth= 4, max_features= 'log2', min_samples_leaf= 5, min_samples_split= 5, n_estimators= 50, subsample= 0.7)
boost.fit(x_train,y_train)
```

```python
b_pred=boost.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

maeb = mean_absolute_error(y_test, b_pred)
mseb = mean_squared_error(y_test, b_pred)
rmseb = np.sqrt(mseb)
r2b = r2_score(y_test, b_pred)

print(f"Mean Absolute Error (MAE): {maeb}")
print(f"Mean Squared Error (MSE): {mseb}")
print(f"Root Mean Squared Error (RMSE): {rmseb}")
print(f"R-squared (RÂ²) Score: {r2b}")
```
